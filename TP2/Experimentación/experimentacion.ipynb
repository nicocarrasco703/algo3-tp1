### Implementación de Kruskal con DSU

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo CSV
df = pd.read_csv("resultados_sin_opt.csv")

# Eliminar el espacio adicional en el nombre del encabezado
df.columns = df.columns.str.strip()

# Obtener los valores del tamaño del input y el tiempo
input_size = df["Tamaño del input"]
time = df["Tiempo (ms)"]

# Crear el gráfico
sns.scatterplot(x=time, y=input_size)
plt.xlabel("Tiempo (ms)")
plt.ylabel("Tamaño del input")

# Mostrar el gráfico
plt.show()
```

### Comparativa

```python
# Leer los datos de los dos archivos CSV
df1 = pd.read_csv("resultados_sin_opt.csv")
df2 = pd.read_csv("resultados_ubs.csv")

# Eliminar el espacio adicional en el nombre del encabezado
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Calcular la diferencia de tiempo
diferencia_tiempo = df2["Tiempo (ms)"] - df1["Tiempo (ms)"]

# Calcular el porcentaje de tiempo adicional
porcentaje_adicional = (diferencia_tiempo / df1["Tiempo (ms)"]) * 100

# Calcular el promedio del porcentaje de tiempo adicional
porcentaje_promedio = porcentaje_adicional.mean()

# Imprimir el porcentaje promedio de tiempo adicional
print("Porcentaje promedio de tiempo adicional:", porcentaje_promedio)
```
